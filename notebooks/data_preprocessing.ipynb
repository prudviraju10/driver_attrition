```python
df_new = df.groupby("Driver_ID",as_index=False).agg({"MMM-YY":"first", "Age":"last", "Gender":"first", "City":"last", "Education_Level":"last", "Income":"mean", "Dateofjoining":"first", "LastWorkingDate":"last", "Joining Designation":"first", "Grade":"last", "Total Business Value":"mean", "Quarterly Rating":"last"})
```

```python
df_new.head()
```

```python
def func(x):
    y = x.values
    # if values are decreased
    if np.any(y != np.sort(y)):
        return 0 
    # for all same values
    if len(np.unique(y)) == 1:
        return 1
    # if values are increased
    return 2
```

```python
df_new["Grade_change"] = df.groupby("Driver_ID", as_index=False)["Grade"].apply(func)["Grade"]
```

```python
df_new["Income_change"] = df.groupby("Driver_ID", as_index=False)["Income"].apply(func)["Income"]
```

```python
df_new["Quaterly_rating_change"] = df.groupby("Driver_ID", as_index=False)["Quarterly Rating"].apply(func)["Quarterly Rating"]
```
